### 1.学生数据分析（总分数：60）

#### 1.数据导入并展示, 文件位置位于0.Teacher/data/下（10分）
   - 请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（5分）   
   - 请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（5分）

```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df_info = pd.read_csv("../../Data/student-info.csv",sep=";")
df_score = pd.read_csv("../../Data/student-score.csv",sep=";")
```

```python
df_info.head()
```

```python
df_score.head()
```

#### 2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (5分）

```python
df = pd.merge(df_info,df_score,on="ID")

df.head()
```

#### 3.把"df"所有的列展示出来(5分)

```python
df.columns
```

#### 4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
df.groupby("sex").agg({"G1":np.mean,"G2":np.mean,"G3":np.mean})
```

#### 5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
df.groupby("age").agg({"G1":np.mean,"G2":np.mean,"G3":np.mean})
```

```python
df.groupby("age").agg({"G1":np.mean,"G2":np.mean,"G3":np.mean}).plot(kind="bar")
```

#### 6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
df.groupby("school").agg({"G1":np.mean,"G2":np.mean,"G3":np.mean})
```

```python
df.groupby("school").agg({"G1":np.mean,"G2":np.mean,"G3":np.mean}).T.plot(kind="bar")
```

#### 7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(10分)

```python
df["score_count"] = df["G1"] + df["G2"] + df["G3"]
df[df["school"] == "GP"].sort_values(["score_count"]).tail(10)
```

```python
df[df["school"] == "MS"].sort_values(["score_count"]).tail(10)
```

   - 请将文件 sales.csv 中的内容读入pandas DataFrame "df_sales"，并展示前五行以及打印df_sales的shape（4分）

```python
df_sales = pd.read_csv("../Data/Walmart/sales.csv")
df_sales.head()
```

```python
df_sales.shape
```

#### 8. 我们发现 df_sales 中每个Store有很多个Dept(department)， 我们只关心整家店的销售情况，所以请把这张表格按照Store和Date来计算整店销售总额，IsHoliday这一列不必保留。结果保存在df_sales_by_store中。（10分）

-2             
我们只关心整家店的销售情况

```python
df_sales_by_store = df_sales.groupby(["Store","Date"]).agg(np.sum).drop(["IsHoliday"],axis=1)
```

```python
# 改正错误后

df_sales_by_store = df_sales.groupby(["Store","Date"]).agg({"Weekly_Sales":sum})
df_sales_by_store.head()
```

#### 9. 使用 Store 和 Date 这两列合并 df_sales_by_store 和 df_feats 两张表格，结果保存在 df_sale_feats 。（10分）

```python
df_sale_feats = pd.merge(df_sales_by_store,df_feats,on=["Store","Date"])
```

```python
df_sale_feats.head()
```

#### 10. 使用 Store 这一列合并 df_sale_feats 和 df_stores 两张表格，结果仍然保留在df_sale_feats 中。（10分）

```python
df_sale_feats = pd.merge(df_sale_feats,df_stores,on="Store")
```

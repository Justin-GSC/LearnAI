```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

### 1.航班乘客变化分析 (10分)

- 载入相应绘图库及航班乘客数据（flights）

```python
flights_df=sns.load_dataset("../data/flights")
flights_df.head()
```

- 分析年度乘客总量变化情况（折线图）

```python
data = flights_df.groupby('year').sum().reset_index()
```

```python
sns.relplot(x = "year",y = "passengers",data=data,kind="line")
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
sns.barplot(x = "month",y="passengers",data=flights_df)
```

### 2.鸢尾花花型尺寸分析

- 载入iris数据集

```python
iris_df=sns.load_dataset("iris")
iris_df.head()

iris_df["sepal"] = iris_df["sepal_length"] * iris_df["sepal_width"]
iris_df["petal"] = iris_df["petal_length"] * iris_df["petal_width"]
```

- 萼片（sepal）和花瓣（petal）的大小关系（散点图）

```python
sns.relplot(x = "sepal",y = "petal",data = iris_df,hue="species")
```

- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）

```python
sns.barplot(x="species",y="sepal",data=iris_df)
```

```python
sns.barplot(x="species",y="petal",data=iris_df)
```

- 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
sns.pairplot(iris_df.iloc[:,:-2],hue="species")
```

### 3.餐厅小费情况分析
* 载入tips数据集

```python
tips_df=sns.load_dataset("../data/tips")
tips_df.head()
```

- 小费和总消费之间的关系（散点图）

```python
sns.relplot(x="total_bill",y="tip",data=tips_df)
```

- 男性顾客和女性顾客，谁更慷慨（分类箱式图）

```python
sns.catplot(x="sex",y="tip",data=tips_df,kind="box")
```

- 抽烟与否是否会对小费金额产生影响（分类箱式图）

```python
sns.catplot(x="smoker",y="tip",data=tips_df,kind="box")
```

- 工作日和周末，什么时候顾客给的小费更慷慨（分类箱式图）

```python
tips_df["day"] = tips_df["day"].apply(lambda x : "weekday" if x in ["Sat","Sun"] else "restday" )
sns.catplot(x="day",y="tip",data=tips_df,kind="box")
```

- 午饭和晚饭，哪一顿顾客更愿意给小费（分类箱式图）

```python
sns.catplot(x="time",y="tip",data=tips_df,kind="box")
```

- 就餐人数是否会对慷慨度产生影响（分类箱式图）

```python
sns.catplot(x="size",y="tip",data=tips_df,kind="box")
```

- 性别+抽烟的组合因素对慷慨度的影响（分组柱状图）

```python
sns.catplot(x="sex",y="tip",hue="smoker",data=tips_df,kind="bar")
```

### 4.泰坦尼克号海难幸存状况分析

* 载入数据集

```python
titanic_df=sns.load_dataset("../data/titanic")
titanic_df.head()
```

* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）

```python
titanic_df["total"] = 1
sns.barplot(x = "pclass",y = "total",data=titanic_df,color = "r")
sns.barplot(x = "pclass",y = "survived",data=titanic_df,color = "b")
```

* 不同性别的幸存比例（堆积柱状图）

```python
titanic_df["total"] = 1
sns.barplot(x = "sex",y = "total",data=titanic_df,color = "r")
sns.barplot(x = "sex",y = "survived",data=titanic_df,color = "b")
```

* 幸存和遇难乘客的票价分布（分类箱式图）

```python
sns.catplot(x="survived",y="fare",data=titanic_df,kind="box")
```

* 幸存和遇难乘客的年龄分布（分类箱式图）

```python
sns.catplot(x="survived",y="age",data=titanic_df,kind="box")
```

* 不同上船港口的乘客仓位等级分布（分组柱状图）

```python
sns.barplot(x="class",y="survived",hue="embarked",data=titanic_df)
```

* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）

```python
sns.catplot(x="survived",y="sibsp",data=titanic_df,kind="box")
```

* 幸存和遇难乘客父母子女的数量分布（分类箱式图）

```python
sns.catplot(x="survived",y="parch",data=titanic_df,kind="box")
```

* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
sns.countplot(x="survived",hue="alone",data=titanic_df)
```

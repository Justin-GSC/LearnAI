## 练习1：航班乘客变化分析
* 分析年度乘客总量变化情况（折线图）
* 分析乘客在一年中各月份的分布（柱状图）

```python
data = sns.load_dataset("flights")
data.head()
# 年份，月份，乘客数
```

```python
sns.relplot(x="year",y="passengers",data=data,kind="line",ci=None)
```

```python
sns.catplot(x="month",y="passengers",data=data,kind="bar")
```

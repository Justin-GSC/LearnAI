# Pandas 内置的plot

> plot() 中 figsize=(2,5) 调整大小   alpha = 0.2 调整透明度

```python
import pandas as pd 
import numpy as np
%matplotlib inline
```

## 折线图

```python
nvda = pd.read_csv("../Data/NVDA.csv",index_col=0,parse_dates=[0])
nvda["Open"].plot()
```

df.plot(kind = )

* bar/barh  条形图 
* hist 直方图 
* boxplot 盒形图 
* area 面积 
* scatter 散点图  

## 条形图&堆积条形图

- 堆积条形图 stacked=True

```python
nvda.iloc[:10,:4].plot.bar()
```

```python
nvda.iloc[:10,:4].plot.bar(stacked = True)
```

```python
nvda.iloc[:10,:4].plot.barh(stacked = True)
```

```python
nvda.iloc[:10,:4].plot.barh()
```

## 直方图

```python
nvda.iloc[:10,:4].plot.hist(bins=30 ) # bins 分组数量
```

### 为每列绘制直方图

```python
nvda.iloc[:10,:4].hist()  #直接使用hist
```

## 箱型图/盒形图/箱线图

> Boxplot 可以绘制调用Series.box.plot()和DataFrame.box.plot()或DataFrame.boxplot()来可视化每列中的值得分布

```python
nvda.iloc[:10,:4].plot.box()
```

## 区域块图形

> 可以使用Series.plot.area()或DataFrame.plot.area()方法创建

```python
nvda.iloc[:10,:4].plot.area()
```

## 散点图形

```python
df = nvda.iloc[:100,2:8]
df.head()
```

```python
df.plot.scatter(x="Low",y = "Adj Close")
```

## 饼状图

> 使用DataFrame.plot.pie()创建

```python
df = pd.DataFrame({'mass': [0.330, 4.87 , 5.97],
                   'radius': [2439.7, 6051.8, 6378.1]},
                  index=['Mercury', 'Venus', 'Earth'])
df.plot.pie(y="radius")
```

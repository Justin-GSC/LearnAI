# matplotlib

## 一副可视化图的基本结构

> * Data:数据区，包括数据点，描绘形状
> * Axis:坐标轴，包括X轴，y轴及其标签、刻度
> * Title:标题，数据图的描述
> * Legend:图例，区分图中包含的多种曲线或不同分类的数据
> 
>
> * 其他的还有图形文本(Text)、注解(Annotate)等其他描述

### 导入相关包

> - matplotlib.pyplot
> - pylab
> - numpy

```python
# coding:utf-8
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

### 准备数据

```python
x = np.arange(0,12,0.3)

y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.sqrt(x)
```

### 绘制基本曲线

```python
plt.rcParams["figure.figsize"] = (6,4) #基本图框大小

plt.plot(x,y1,color="blue",linewidth=1.5,linestyle="-",marker=".",label=r'$y = cos(x)$')
plt.plot(x,y2,color="red",linewidth=1.5,linestyle="-",marker="*",label=r'$y = sin(x)$')
plt.plot(x,y3,color="green",linewidth=1.5,linestyle="-",marker="x",label=r'$y = sqrt(x)$')
```

### axis 设置

#### 移动坐标轴

```python
ax = plt.subplot(111) #图的位置，1 * 1 的第一个
# ax1 = plt.subplot(221)
# ax2 = plt.subplot(222)
# ax3 = plt.subplot(223)
# ax4 = plt.subplot(224)# 2 * 2 的第一个

ax.spines["top"].set_color('none')   # 关闭上边框线
ax.spines["right"].set_color('none') # 关闭右边框线


ax.xaxis.set_ticks_position('top')    # 设定 x轴的lebel 在上下边
ax.spines['bottom'].set_position(('data', 0.3)) # 设定边框位置


ax.yaxis.set_ticks_position('right') # 设定 y轴的lebel 在左右边
ax.spines['left'].set_position(('data', 0.3)) # 设定边框位置
```

#### 刻度间隔lim，刻度标签ticks

```python
plt.xlim(x.min()*1.1, x.max()*1.1)  # 设置 x, y 轴的刻度取值范围
plt.ylim(-1.5, 4.0)  # 设置 x, y 轴的刻度标签值

plt.xticks([2, 4, 6, 8, 10], [r'two', r'four', r'six', r'8', r'10'])
plt.yticks([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0],
    [r'-1.0', r'0.0', r'1.0', r'2.0', r'3.0', r'4.0'])
```

#### 设置X、Y坐标轴和标题

```python
# 设置标题、x轴、y轴
plt.title(r'$the \ function \ figure \ of \ cos(), \ sin() \ and \ sqrt()$', fontsize=12 )

plt.xlabel(r'$the \ input \ value \ of \ x$', fontsize=12, labelpad=1)  # fontsize 字体大小 labelpad 距离坐标轴的距离
plt.ylabel(r'$y = f(x)$', fontsize=12, labelpad=1)
```

### 网格线开关

```python
plt.grid(True)
```

### 设置文字描述、注解

```python
plt.text(0.8, 0.9, r'$x \in [0.0, \ 10.0]$', color='k', fontsize=15) # 位置，内容，颜色，大小
plt.text(0.8, 0.8, r'$y \in [-1.0, \ 4.0]$', color='k', fontsize=15)
```

```python
# 特殊点添加注解
plt.rcParams["figure.figsize"] = (8,6)

plt.plot(x,y1,color="blue",linewidth=1.5,linestyle="-",marker=".",label=r'$y = cos(x)$')
plt.plot(x,y2,color="red",linewidth=1.5,linestyle="-",marker="*",label=r'$y = sin(x)$')
plt.plot(x,y3,color="green",linewidth=1.5,linestyle="-",marker="x",label=r'$y = sqrt(x)$')

plt.scatter([8,],[np.sqrt(8),], 50, color ='m')  # 使用散点图放大当前点

plt.annotate(r'$2\sqrt{2}$', xy=(8, np.sqrt(8)), xytext=(8.5, 2.2), fontsize=16, 
             color='#090909', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=0.1', color='#090909'))

plt.legend(loc=2)

# 文字 点的坐标  文字的坐标 大小  颜色  箭头 
```

### 设置图例

> 可使用如下两种方式，给绘图设置图例：
> * 1: 在 plt.plot 函数中添加 label 参数后，使用 plt.legend(loc=’up right’)
> * 2: 不使用参数 label, 直接使用如下命令：plt.legend(\['cos(x)', 'sin(x)', 'sqrt(x)'\], loc='upper right')
>

## 完整的绘制

```python
#coding:utf-8

import numpy as np
import matplotlib.pyplot as plt
from pylab import *


plt.rcParams["figure.figsize"] = (12,8) #基本图框大小

# 定义数据部分
x = np.arange(0., 10, 0.2)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.sqrt(x)

# 绘制 3 条函数曲线
plt.plot(x, y1, color='blue', linewidth=1.5, linestyle='-', marker='.', label=r'$y = cos{x}$')
plt.plot(x, y2, color='green', linewidth=1.5, linestyle='-', marker='*', label=r'$y = sin{x}$')
plt.plot(x, y3, color='m', linewidth=1.5, linestyle='-', marker='x', label=r'$y = \sqrt{x}$')

# 坐标轴上移
ax = plt.subplot(111)
ax.spines['right'].set_color('none')     # 去掉右边的边框线
ax.spines['top'].set_color('none')       # 去掉上边的边框线

# 移动下边边框线，相当于移动 X 轴
ax.xaxis.set_ticks_position('bottom')    
ax.spines['bottom'].set_position(('data', 0))

# 移动左边边框线，相当于移动 y 轴
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

# 设置 x, y 轴的取值范围
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(-1.5, 4.0)

# 设置 x, y 轴的刻度值
plt.xticks([2, 4, 6, 8, 10], [r'2', r'4', r'6', r'8', r'10'])
plt.yticks([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], 
    [r'-1.0', r'0.0', r'1.0', r'2.0', r'3.0', r'4.0'])

# 添加文字
plt.text(0.8, 0.8, r'$x \in [0.0, \ 10.0]$', color='k', fontsize=15)
plt.text(0.8, 0.9, r'$y \in [-1.0, \ 4.0]$', color='k', fontsize=15)

# 特殊点添加注解
plt.scatter([8,],[np.sqrt(8),], 50, color ='m')  # 使用散点图放大当前点
plt.annotate(r'$2\sqrt{2}$', xy=(8, np.sqrt(8)), xytext=(8.5, 2.2), fontsize=16, color='#090909', arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=0.1', color='#090909'))

# 设置标题、x轴、y轴
plt.title(r'$the \ function \ figure \ of \ cos(), \ sin() \ and \ sqrt()$', fontsize=19)
plt.xlabel(r'$the \ input \ value \ of \ x$', fontsize=18, labelpad=88.8)
plt.ylabel(r'$y = f(x)$', fontsize=18, labelpad=12.5)

# 设置图例及位置
plt.legend(loc=2)    
# plt.legend(['cos(x)', 'sin(x)', 'sqrt(x)'], loc='up right')

# 显示网格线
plt.grid(True)    

# 显示绘图
plt.show()
```

* 折线图：matplotlib.pyplot.plot(data)

```python
x = np.arange(-5,5,1)
y = x ** 2
plt.plot(x,y)
```

```python
x = np.random.normal(size=10000)
plt.hist(x, bins=10)
```

```python
plt.rcParams["figure.figsize"] = (8,8)
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
plt.scatter(x,y)
```

#### 箱式图科普
* 上边缘（Q3+1.5\*IQR/“箱子高度”）、下边缘（Q1-1.5\*IQR/“箱子高度”）、IQR=Q3-Q1
* 上四分位数（Q3）、下四分位数（Q1）
* 中位数
* 异常值
* 处理异常值时与3$\sigma$标准的异同：统计边界是否受异常值影响、容忍度的大小

* 关联分析、数值比较：散点图、曲线图
* 分布分析：灰度图、密度图
* 涉及分类的分析：柱状图、箱式图

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
```

# seaborn

> seaborn 的API 设计非常清晰，适合新手等操作。而且其内置的数据分析功能，不需要对数据进行过多的处理即可画出可视化图形

[seaborn API](https://seaborn.pydata.org/api.html#relational-api)

## Seaborn 的画布设置

```python
plt.rcParams["figure.figsize"] = (6,8)
```

## Relational plots  绘制关系图

> 通过relational plots 看 x,y 之间的关系

### [sns.relplot([x, y, hue, size, style, data, row, …])](https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot)

```python
tips = sns.load_dataset("tips")

sns.relplot(x = "total_bill",y = "tip",col="day",col_wrap=2,data=tips)
```

### [sns.scatterplot([x, y, hue, size, style, data, row, …])](https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot)

```python
plt.close("all")
index = pd.date_range("1 1 2000", periods=100,
                       freq="m", name="date")
data = np.random.randn(100, 4).cumsum(axis=0)
wide_df = pd.DataFrame(data, index, ["a", "b", "c", "d"])
ax = sns.scatterplot(data=wide_df)
```

### [sns.lineplot([x, y, hue, size, style, data, row, …])](https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot)

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
ax = sns.lineplot(x="time", y="firing_rate",
                   hue="coherence", style="choice",
                  data=dots)
```

## Categorical plots

### [sns.catplot([x, y, hue, data, row, col, …])](https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot)

```python
exercise = sns.load_dataset("exercise")
g = sns.catplot(x="time", y="pulse", hue="kind", data=exercise)
```

### [sns.stripplot([x, y, hue, data, order, …])](https://seaborn.pydata.org/generated/seaborn.stripplot.html#seaborn.stripplot)

```python
sns.stripplot(x="sex", y="total_bill", hue="day",
                 data=tips, jitter=True)
```

### [sns.swarmplot([x, y, hue, data, order, …])](https://seaborn.pydata.org/generated/seaborn.swarmplot.html#seaborn.swarmplot)

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

### [sns.boxplot([x, y, hue, data, order, hue_order, …])](https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn.boxplot)

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

### [sns.violinplot([x, y, hue, data, order, …])](https://seaborn.pydata.org/generated/seaborn.violinplot.html#seaborn.violinplot)

```python
sns.violinplot(x="time", y="tip", data=tips,
                   order=["Dinner", "Lunch"])
```

### [sns.boxenplot([x, y, hue, data, order, …])](https://seaborn.pydata.org/generated/seaborn.boxenplot.html#seaborn.boxenplot)

```python
sns.boxenplot(x="day", y="total_bill", data=tips)
```

### [sns.pointplot([x, y, hue, data, order, …])](https://seaborn.pydata.org/generated/seaborn.pointplot.html#seaborn.pointplot)

```python
sns.pointplot(x="tip", y="day", data=tips, join=False)
```

### [sns.barplot([x, y, hue, data, order, hue_order, …])](https://seaborn.pydata.org/generated/seaborn.barplot.html#seaborn.barplot)

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
```

### [sns.countplot([x, y, hue, data, order, …])](https://seaborn.pydata.org/generated/seaborn.countplot.html#seaborn.countplot)

```python
sns.set(style="darkgrid")
titanic = sns.load_dataset("titanic")
sns.countplot(x="class", hue="who", data=titanic)
```

## Distribution plots

### [sns.joinplot(x, y[, data, kind, stat_func, …])](https://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot)

```python
sns.jointplot("total_bill", "tip", data=tips,kind="reg")
```

### [sns.pairplot(data[, hue, hue_order, palette, …])](https://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot)

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

### [sns.distplot(a[, bins, hist, kde, rug, fit, …])](https://seaborn.pydata.org/generated/seaborn.distplot.html#seaborn.distplot)

```python
x = np.random.normal(0,0.1,100)
sns.distplot(x)
```

### [sns.kdeplot(data[, data2, shade, vertical, …])](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot)

```python

sns.set(color_codes=True)
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov, size=50).T
ax = sns.kdeplot(x,y,shade=True)
```

### [sns.rugplot(a[, height, axis, ax])](https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot)

```python
a = np.array([1,1,1,2,3,1,2,3,5,66,4])
sns.rugplot(a)
```

## Regression plots

### [sns.lmplot(x, y, data[, hue, col, row, palette, …])](https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot)

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
               palette=dict(Yes="g", No="m"))
```

### [sns.regplot(x, y[, data, x_estimator, x_bins, …])](https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot)

```python
sns.regplot(x="size", y="total_bill", data=tips,
                 x_estimator=np.mean, logx=True, truncate=True)
```

### [sns.residplot(x, y[, data, lowess, x_partial, …])](https://seaborn.pydata.org/generated/seaborn.residplot.html#seaborn.residplot)

```python
sns.residplot(x="total_bill", y="tip",data=tips)
```

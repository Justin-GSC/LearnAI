#### 1.简述Hadoop解决了单机的什么问题？Spark与之相比又有哪些优点？

【参考答案】
#### hadoop是一个适合大数据的分布式存储和计算的平台，相比单机，它具有如下优点：
- 高效性:这也是hadoop的核心竞争优势所在，接受到客户的数据请求后，hadoop可以在数据所在的集群节点上并发处理。   
- 可靠性:通过分布式存储，hadoop可以自动存储多份副本，当数据处理请求失败后，会自动重新部署计算任务。   
- 扩展性:hadoop的分布式存储和分布式计算是在集群节点完成的，这也决定了hadoop可以扩展至更多的集群节点。   
- 容错性:hadoop能够自动保存数据的多个副本，并且能够自动将失败的任务重新分配   

#### Spark相比hadoop的优点有：
-  丰富的API，而后者只有map和reduce。
-  中间结果不存磁盘，而后者需要把中间结果写磁盘。
-  线程池模型减少task启动开销，而后者任务调度和启动开销很大。
-  能充分利用内存速度快优势，减少IO操作，而后者不能。
-  可以避免排序操作，而hadoop中map和reduce都需要排序。
-  适合迭代计算（机器学习算法），而后者不适合。

#### 3.请简述DataFrame的含义，并写出针对DataFrame的两类操作（对应RDD的transformation与action),每类下至少三种的操作。

【参考答案】

- 含义：DataFrame是Spark SQL中的数据模型，它是一个基于RDD实现的由多个列组成的结构化的分布式数据集合，相当于关系数据库中的一张表。
- 操作
    - transformation
        - filter() 传入筛选条件表达式，得到DataFrame类型的返回结果
        - groupBy() 对记录进行分组
        - distinct() 返回一个不包含重复记录的DataFrame
    - action
        - count() 返回dataframe的列数
        - collect() 获取所有数据到数组
        - first() 获取第一行记录
        - take(n) 获取前n行数据 
        - takeAsList(n) 按list的方式返回dataframe的前n行

##### 2.数据处理1   (5分)
- 对以上数据进行清洗，把记录中带的引号去除掉，得到content_rdd
- tip：参考map函数的用法

你将得到类似以下结果的RDD：
```
[
[u'date', u'time', u'size', u'r_version', u'r_arch', u'r_os', u'package', u'version', u'country', u'ip_id'], 
[u'2015-12-12', u'13:42:10', u'257886', u'3.2.2', u'i386', u'mingw32', u'HistData', u'0.7-6', u'CZ', u'1'], 
[u'2015-12-12', u'13:24:37', u'1236751', u'3.2.2', u'x86_64', u'mingw32', u'RJSONIO', u'1.3-0', u'DE', u'2'],

```

````python
content_rdd=raw_content.map(lambda x:x.replace('"','')).map(lambda x:x.split(','))
content_rdd.take(5)
````

##### 3. 数据处理2（5分)
- 有如下的text，请使用flatmap得到包含所有字幕的list
- tip：请参考课程flatmap的使用，特别注意这里不同分割符的处理

text=["a b c", "d e", "f#g#h", "m n q", "r,q,w", "j%r%q"]

你将得到如下结果的RDD：
```
[a, b, c, d, e, f, g, h, m, n, q, r, q, w, j, r, q]
```

````python
import re
regexp = r'[\s#,%]'
reg = re.compile(regexp)
def remove_symbol(s):
    s = reg.sub('-',s)
    return s.split('-')
    
text=["a b c", "d e", "f#g#h", "m n q", "r,q,w", "j%r%q"]
text_rdd = sc.parallelize(text)
text_rdd = text_rdd.flatMap(lambda s:remove_symbol(s))
print(text_rdd.collect())
````

##### 4. 数据处理3（5分）
- 从刚才的content_rdd中取出第7列，对不同的package类别进行统计计数
- tip：可以使用map reduce或者pair-rdd reduceByKey



````python
package_rdd = content_rdd.map(lambda s:s[6]).filter(lambda s:s!='package').cache()
package_count_rdd = package_rdd.map(lambda s:(s,1)).reduceByKey(lambda x,y:x+y)
package_count_rdd.take(5)
````

##### 5.排序   (5分)
- 取出数量最多的10个package和出现的频次
- tip：注意sortbykey的使用

你将看到类似如下的结果：
```
[(4783, u'Rcpp'),
 (3913, u'ggplot2'),
 (3748, u'stringi'),
 (3449, u'stringr'),
 (3436, u'plyr'),
 (3265, u'magrittr'),
 (3223, u'digest'),
 (3205, u'reshape2'),
 (3046, u'RColorBrewer'),
 (3007, u'scales')]
```

````python
package_count_sort_rdd = package_count_rdd.map(lambda t:(t[1],t[0])).sortByKey(ascending=False).cache()
package_count_sort_rdd.take(10)
````

##### 6.过滤  (5分)

- 取出top5的package对应的数据记录
- tip：注意filter的使用

````python
top5_package = [tup[1] for tup in package_count_sort_rdd.take(5)]
top5_package_rdd = content_rdd.filter(lambda s:s[6] in top5_package).cache()
top5_package_rdd.take(5)
````

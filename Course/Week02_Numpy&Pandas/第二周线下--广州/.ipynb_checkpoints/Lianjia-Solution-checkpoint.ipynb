```python
#  Index, Datatype and Memory information
lj_data.info()
```

```python
# 房龄字符串清洗,回顾(IPO 例子)

lj_data = lj_data.applymap(lambda x:x if not '年建板楼' in str(x) else x.replace('年建板楼', ''))
lj_data = lj_data.applymap(lambda x:x if not '年建板塔结合' in str(x) else x.replace('年建板塔结合', ''))
lj_data = lj_data.applymap(lambda x:x if not '年建塔楼' in str(x) else x.replace('年建塔楼', ''))
lj_data = lj_data.applymap(lambda x:x if not '年建' in str(x) else x.replace('年建', ''))
lj_data = lj_data.applymap(lambda x:x if not '板楼' in str(x) else x.replace('板楼', ''))
lj_data['build_time']
```

```python
lj_data['build_time'].notnull()
lj_data[lj_data['build_time'].notnull()].sort_values(['build_time']).tail(20)['view_num'].mean()
```

### 在租房屋的建筑类型有哪些

```python
import re
lj_data['build_type'] = lj_data['build_time'].map(lambda x:re.findall('\d{4}年建(.+)?',str(x),re.I))
lj_data['build_type'] = lj_data['build_type'].map(lambda x:''.join(x))
lj_data
```

### 哪个城区的租房价格最贵，平均价格是多少

```python
import numpy as np
lj_data.groupby('district')['price'].agg([np.mean, np.max, np.min])
```

### 不同房型的平均/最大/最小面积

```python
lj_data.groupby('house_type')['area'].agg([np.mean, np.min, np.max])
```

### 哪个地铁口附近的房子最多

```python
lj_data['dt'] = lj_data['extra_info'].map(lambda x:re.findall('距离(.+)站', x, re.I))
lj_data['dt'] = lj_data['dt'].map(lambda x:''.join(x))
lj_data[lj_data['dt'] == ''] = 'None'

lj_data.groupby('dt')['district'].count()
```

### 地铁附近的房子平均价格 比 非地铁的高多少

```python
# 地铁附近的平均价格
lj_data[lj_data['dt']!='None']['price'].mean()
lj_data
```

## IPO 数据分析

#### 数据列说明
- Date 日期
- Issuer 发行者
- Symbol 股票代码
- Manager 承销商
- OfferPrice 发行价
- OpeningPrice 开盘价
- DayClose 首日收盘价
- ChangeDay 首日收盘价格较发行价变化率
- ChangeOpening 首日开盘价较发行价变化（绝对金额）
- ChangeClose 首日收盘价较发行价变化（绝对金额）
- StarRatings 评级


#### 画出首日收盘价格较发行价变化率的分布


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
ipos=pd.read_csv('./ipos.csv',encoding='latin-1')
ipos.head(20) # 默认打印前 5 行 head(n)
```

```python
#数据清洗1 使得字符串标准化
print(ipos['Date'][1080])
ipos['Date'][1080]='2012/11/20'
ipos=ipos.applymap(lambda x:x if not '$' in str(x) else x.replace('$',''))
ipos=ipos.applymap(lambda x:x if not '%' in str(x) else x.replace('%',''))
ipos=ipos.applymap(lambda x:x if not '(' in str(x) else x.replace('(',''))
ipos=ipos.applymap(lambda x:x if not ')' in str(x) else x.replace(')',''))
ipos.head(20)
```

```python
#数据清洗2
ipos.replace('N/C',0,inplace=True)
# 日期格式化
ipos['Date']=pd.to_datetime(ipos['Date'])
# 字符串转 float
ipos['OfferPrice']=ipos['OfferPrice'].astype('float')
ipos['OpeningPrice']=ipos['OpeningPrice'].astype('float')
ipos['DayClose']=ipos['DayClose'].astype('float')
ipos['ChangeDay']=ipos['ChangeDay'].astype('float')
ipos['ChangeClose']=ipos['ChangeClose'].astype('float')
ipos['ChangeOpening']=ipos['ChangeOpening'].astype('float')
ipos['StarRatings']=ipos['StarRatings'].astype('float')
ipos.head(20)
```

```python
ipos['Date'].dt.year
```

```python
ipos.groupby(ipos['Date'].dt.year)['ChangeDay'].median()
```

- pandas提供了一个灵活高效的 groupby 功能，它使你能以一种自然的方式对数据集进行切片、切块、摘要等操作。
- 根据一个或多个键（可以是函数、数组或 DataFrame 列名）拆分 pandas 对象。计算分组摘要统计，如计数、平均值、标准差，或用户自定义函数。对 DataFrame 的列应用各种各样的函数。
- 应用组内转换或其他运算，如规格化、线性回归、排名或选取子集等。计算透视表或交叉表。执行分位数分析以及其他分组分析。

```python
#按年比较首日收盘价格较发行价变化率的中位数
ipos.groupby(ipos['Date'].dt.year)['ChangeDay'].median().plot(kind='bar',
                   figsize=(10, 5),
                   color='b',
                   title='1st day median change')
```

```python
# 数值观察首日收盘价格较发行价变化率的大体情况
# 一次性输出多个描述性统计指标
ipos['ChangeDay'].describe(include='all')
```

```python
#画出首日收盘价格较发行价变化率的分布
ipos['ChangeDay'].hist(figsize=(6,4),bins=100,color='g')

#请描述这个分布(近似高斯分布)
```

```python
#想知道，如果对新股以开盘价买入，收盘价卖出这样策略的收益 情况？

ipos['CloseMinusOpen']=ipos['DayClose']-ipos['OpeningPrice']
ipos['CloseMinusOpenRate']=(ipos['CloseMinusOpen']/ipos['OpeningPrice'])*100
ipos['CloseMinusOpenRate'].hist(figsize=(6,4),bins=100,color='g')
```

```python
#再想想，如果是以IPO发价购买，首日开盘价卖出这样策略呢？
ipos['IpoProfit']=ipos['OpeningPrice']-ipos['OfferPrice']
ipos['IpoProfit'].hist(bins=250,figsize=(10,10))
ipos['IpoProfit'].describe()
```

```python
#想知道在2009年以后，我们按开盘买入，收盘卖出的策略去执行的情况
ipos[ipos['Date']>'2009-1-1']['ChangeDay'].describe()
```
